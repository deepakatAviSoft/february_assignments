```python
# Replace "-" with NaN
df.replace('-', pd.NA, inplace=True)

# Find null values in columns
null_counts = df.isnull().sum()

# Print the count of null values in each column
print(null_counts)
```

```python
# Calculate the percentage of null values in each column
null_percentages = (df.isnull().sum() / len(df)) * 100
null_percentages = round(null_percentages)
# Print the percentage of null values in each column
print(null_percentages)
```

```python
# Get the data type of all columns
column_types = df.dtypes

# Print the data type of each column
print(column_types)
```

```python
# Find the median value of each column
medians = df.median()

# Print the median value of each column
print(medians)
```
